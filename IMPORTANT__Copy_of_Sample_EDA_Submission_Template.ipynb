````python
# Checking for outliers
print(df.availability_365.describe())
plt.figure(figsize=(12,8))
sns.histplot(data=df, x='availability_365', bins=100,kde=True)
plt.title("Histogram of avaibility - days per year",color='red')
plt.xlabel("availability")
plt.ylabel("Frequency")
plt.show()
````

````python
# BOXPLOT OF AVAILABILITY_365
plt.figure(figsize=(12,8))
sns.boxplot(
data=df,
x='availability_365'
)
plt.title('Boxplot of Availability (Days Per Year)', color='red')
plt.xlabel('Availability')
plt.show()
````

Insights Found:

-Availability distribution is highly uneven. Some listings remain available throughout most of the year.

````python
#HANDLING OUTLIERS IN availability_365
# Capping extreme availability values using 99th percentile
availability_upper = df['availability_365'].quantile(0.99)
# Applying clipping
df['availability_365'] = np.where(
df['availability_365'] > availability_upper,
availability_upper,
df['availability_365']
)
print(df['availability_365'].describe())
````

````python

#MINIMUM NIGHTS BOXPLOT
print(df.minimum_nights.describe())

plt.figure(figsize=(10, 4))
sns.boxplot(x=df['minimum_nights'])
plt.title("Boxplot of Minimum Nights")
plt.grid(True, axis='x', linestyle='--', alpha=0.5)
plt.show()
````

Insights Found:

Extreme outliers exist in minimum stay requirements.

````python
#MINIMUM NIGHTS DISTRIBUTION

# Optional: Filter out extreme outliers for clearer plot
filtered = df[df['minimum_nights'] <= 30]

# Create subplots: 2 rows, 1 column
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(10, 6), gridspec_kw={'height_ratios': [1, 3]})

# Boxplot (top)
sns.boxplot(x=filtered['minimum_nights'], ax=axes[0], color='skyblue')
axes[0].set(title='Box Plot of Minimum Nights')
axes[0].set(xlabel='')

# Histogram (bottom)
sns.histplot(filtered['minimum_nights'], bins=30, ax=axes[1], color='salmon')
axes[1].set(title='Histogram of Minimum Nights', xlabel='Minimum Nights', ylabel='Count')

plt.tight_layout()
plt.show()
````

````python
# HANDLING PRICE OUTLIERS
print(df.price.describe())
plt.figure(figsize=(10, 4))
sns.histplot(x=df['price'],bins=100,kde=True)
plt.title("Hitogram  of Price")
plt.grid(True, axis='x', linestyle='--', alpha=0.5)
plt.show()
````

Insight

Presence of Severe Outliers, due to luxury listings

````python
#HANDLING PRICE OUTLIERS
# Checking price distribution
print(df['price'].describe())
# Capping price outliers using 99th percentile
price_upper = df['price'].quantile(0.99)
# Applying clipping
df['price'] = np.where(df['price'] > price_upper, price_upper, df['price'])
````

````python
print(df['price'].describe())
````

````python
# Creating new columns for better analysis
# Creating price categories
df['price_category'] = pd.cut(df['price'], bins=[0,100,300,1000], labels=['Budget', 'Mid-Range', 'Luxury'])
# Creating host categories
df['host_type'] = np.where(df['calculated_host_listings_count'] > 1, 'Professional Host',
'Individual Host')
# Creating availability categories
df['availability_status'] = pd.cut(df['availability_365'], bins=[-1,0,120,365],
labels=['Unavailable', 'Low Availability', 'High Availability'])
````
